# Data Visulizations

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
plt.hist(data['TARGET_deathRate'],bins=30,color = 'red',
        edgecolor ='black')
plt.xlabel('Death Rate')
plt.ylabel('Frequency')
plt.title('Histogram of Death Rate')
```

## Keypoints from the Histogram 
1. We can observe outliners in the dataset. 
2. The data looks like right skewed (slightely positively skewed).
3. Mean is slightly greater than median.
4. There are 3 peaks in the data.

```python
plt.figure(figsize =(8,3))
bp = plt.boxplot(data['TARGET_deathRate'],patch_artist = True,vert=0)
colors = ['#00FF00','#0F00FF', '#F00FF0','#FFFF0F']
for patch in bp['boxes']:  
    patch.set_facecolor('#00FF00')  
for whisker in bp['whiskers']:  
    whisker.set(color ='#8E008B',linewidth = 1.4,linestyle =":")  
for cap in bp['caps']:  
    cap.set(color ='#8E008B',linewidth = 2.1)  
for median in bp['medians']:  
    median.set(color ='blue',linewidth = 3)  
for flier in bp['fliers']:  
    flier.set(marker ='D',color ='#d7298c',alpha = 0.6)   
```

# Keypoints from the Boxplot 
1. We can conform that there are outliners in the dataset.
2. This oultilners will have huge impact while training a model.

```python
avg_ann_count = data['avgAnnCount'].mean()
std_ann_count = data['avgAnnCount'].std()
print("Mean of avganncount :",avg_ann_count)
print("Standard deviation of avganncount :",std_ann_count)


plt.hist(data['avgAnnCount'],bins=20,color='red',edgecolor='black')
plt.xlabel('avgAnnCount')
plt.ylabel('Frequency')
plt.title("Histogram of avganncount")
plt.show()

plt.boxplot(data['avgAnnCount'],meanline = True,widths=0.5,vert=False)
plt.title('Boxplot of avganncount')
plt.show()

corr= data['avgAnnCount'].corr(data['TARGET_deathRate'])
print("The Correlation between the avganncount and target death rate is ",corr)


plt.scatter(data['avgAnnCount'],data['TARGET_deathRate'])
plt.xlabel('avganncount')
plt.ylabel('target_death rate')
plt.title('scatter plot')
plt.show()
```

# Key interpretations

1. The mean of avganncount is 606.33, which means that in each county there are anually 606 cancer diagnosed patients.

2. The Standard deviation of avganncount is 1416.35, tells us how that the data is quite spread out.
In some counties there may be less cancer diagnosis and there may be high cancer diagnosis annually.

3. The correlation between avganncount and target deathrate tell us that if the avganncount incereases then the deathrate decreases(shows negative linear relationship)

```python
from scipy.stats import normaltest

num_col = data.select_dtypes(include=np.number).columns
gaussian_col = []
non_gaussian_col  = []
for col in num_col:
    stat,p = normaltest(data[col])
    alpha = 0.05
    if p>alpha:
        gaussian_col.append(col)
    else:
        non_gaussian_col.append(col)
print(gaussian_col)     
```

```python
len(non_gaussian_col)
```

```python
fig,axes = plt.subplots(1,1)
for i,col in enumerate(gaussian_col):
    data[col].plot(kind='density',ax=axes,subplots=True)
plt.tight_layout()
plt.show()
```
